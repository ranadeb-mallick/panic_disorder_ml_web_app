# Transforming the columns according to need

```python
cols = ['Gender','Family History','Personal History','Current Stressors','Symptoms','Severity','Impact on Life','Demographics',
        'Medical History','Psychiatric History','Substance Use','Coping Mechanisms','Social Support','Lifestyle Factors']

for i,col in enumerate(cols, start=1):
    unique_values=df[col].unique()
    print(i,'. ',col,':-')
    print(unique_values, '\n')
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler
from sklearn.compose import ColumnTransformer, make_column_transformer
from sklearn.pipeline import Pipeline, make_pipeline
```

#  Column Transformations 

```python
tnf = ColumnTransformer(transformers=[
    ('tnf1', OneHotEncoder(sparse_output=False, drop='first', handle_unknown='ignore'),[1, 2, 3, 5, 8, 12, 14]),
    ('tnf2', OrdinalEncoder(categories=[['Low', 'Moderate', 'High']]), [4]),
    ('tnf3', OrdinalEncoder(categories=[['Mild', 'Moderate', 'Severe']]), [6]),
    ('tnf4', OrdinalEncoder(categories=[['Mild', 'Moderate', 'Significant']]), [7]),
    ('tnf5', OrdinalEncoder(categories=[['None', 'Asthma', 'Heart disease', 'Diabetes']]), [9]),
    ('tnf6', OrdinalEncoder(categories=[['None', 'Anxiety disorder', 'Depressive disorder', 'Bipolar disorder']]), [10]),
    ('tnf7', OrdinalEncoder(categories=[['None', 'Alcohol', 'Drugs']]), [11]),
    ('tnf8', OrdinalEncoder(categories=[['Low', 'Moderate', 'High']]), [13]),
], remainder='passthrough')
```

```python
tnf.fit_transform(x_train).shape
```

```python
tnf.transform(x_test).shape
```

```python
from sklearn import set_config
set_config(display='diagram')
```

```python
tnf
```

# Applying standard scalar

```python
scaler = ColumnTransformer(transformers=[
    ('scale', StandardScaler(),slice(0,21))
])
```

```python
scaler
```

# Model: 1 Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression

model1 = LogisticRegression(max_iter=1000)
```

```python
# Creating pipeline 1
pipe1 = make_pipeline(tnf, scaler, model1)
```

```python
pipe1.fit(x_train, y_train)
```

```python
y_pred_lr = pipe1.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

acc_lr = accuracy_score(y_test, y_pred_lr)
print('Accuracy of Logistic Regression = ', acc_lr)
```

```python
cm = confusion_matrix(y_test, y_pred_lr)
print(cm)
```

```python
cr = classification_report(y_test, y_pred_lr)
print(cr)
```

```python
sns.heatmap(cm, annot=True, cmap='PiYG')
plt.xlabel('Predict')
plt.ylabel('Truth')
plt.title('Visualization of confusion metrics')
plt.show()
```
